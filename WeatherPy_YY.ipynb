## Analysis

* Temperatures are the warmest around 20 degree latitude and gets colder in either direction in latitude.
* The hottest temperatures are around 30 degrees latitude.
* Cloudiness seems to be evenly spread out for all latitudes.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

import api_keys

from citipy import citipy

# output_data_file = "output_data/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1680)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1680) 
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

numCities = len(cities)
print(numCities)
```

## Perform API Calls

```python
api_key = api_keys.api_key
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
```

```python
city_name = []
country = []
clouds = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind = []

counter = 1

print(f"Beginning Data Retrieval")
print("-"*20)

for city in cities:
    query_url = f"{url}appid={api_key}&q={city}&units={units}"
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    print(f"Processing Record {counter} of {numCities} | {city} ")
    print(str(query_url))
    counter = counter + 1
    
    try: 
        lng.append(weather_json['coord']['lon'])
        lat.append(weather_json['coord']['lat'])
        clouds.append(weather_json['clouds']['all'])
        date.append(weather_json['dt'])
        humidity.append(weather_json['main']['humidity'])
        max_temp.append(weather_json['main']['temp_max'])
        wind.append(weather_json['wind']['speed'])
        country.append(weather_json['sys']['country'])
        city_name.append(city)
    except KeyError:
        print(f"Info not found for city: {city}")
        
print("Finished Extracting Data")
```

```python
print(len(city_name))
print(len(country))
print(len(lng))
print(len(lat))
print(len(clouds))
print(len(date))
print(len(humidity))
print(len(max_temp))
print(len(wind))
```

```python
city_weather_df = pd.DataFrame({ 
    "city_name": city_name,
    "country_name": country,
    "longitude": lng,
    "latitude": lat,
    "clouds": clouds,
    "date": date,
    "humidity": humidity,
    "max temp": max_temp,
    "wind": wind
})
city_weather_df.head()
```

### Save both a CSV of all data retrieved and png images for each scatter plot.

```python
city_weather_df.to_csv("CSV/city_weather.csv", mode='w')
```

## Scatter Plots

```python
plot_date = time.strftime("%m/%d/%Y")
```

### Temperature (F) vs. Latitude

```python
latitude = city_weather_df['latitude']
temperature = city_weather_df['max temp']
plt.scatter(latitude, temperature, edgecolor="black", linewidths=1, marker="o", alpha=0.7)

plt.title("Temperature (F) vs. Latitude - " + "(" + plot_date +")")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("Figures/plot_001_tempVSlat.png")

plt.show()
```

### Humidity (%) vs. Latitude

```python
humidity_plot = city_weather_df['humidity']
plt.scatter(latitude, humidity_plot, edgecolor="black", linewidths=1, marker="o", alpha=0.7)

plt.title("Humidity (%) vs. Latitude - " + "(" + plot_date +")")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("Figures/plot_002_humidVSlat.png")

plt.show()
```

### Cloudiness (%) vs. Latitude

```python
cloud_plot = city_weather_df['clouds']
plt.scatter(latitude, cloud_plot, edgecolor="black", linewidths=1, marker="o", alpha=0.7)

plt.title("Cloudiness (%) vs. Latitude - " + "(" + plot_date +")")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("Figures/plot_003_cloudVSlat.png")

plt.show()
```

```python
wind_plot = city_weather_df['wind']
plt.scatter(latitude, wind_plot, edgecolor="black", linewidths=1, marker="o", alpha=0.7)

plt.title("Wind Speed (mph) vs. Latitude - " + "(" + plot_date +")")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("Figures/plot_004_windVSlat.png")

plt.show()
```
